```python
# import
import pandas as pd
```

```python
# load data
rookies_df = pd.read_csv('data/nba_hof_rookies.csv')
rookies_df.head()
```

```python
# check types
rookies_df.dtypes
```

```python
# rename some columns
rookies_df.rename(columns={'Name':'name',
                           'Hall of Fame Class':'hof_class',
                           'Year Drafted':
                           'year_drafted',
                           '3P Made':'3PM'},
                  inplace=True)
rookies_df.head()
```

```python
# fill NaN with zeros
rookies_df.fillna(0, inplace=True)

# Change HOF induction year with ones
rookies_df['hof_class'].values[rookies_df['hof_class'].values > 0] = 1
rookies_df.head()
```

```python
# drop columns that are not needed
rookies_df.drop(['name', 'year_drafted'], axis=1, inplace=True)

# dropping for now until I take time to convert from object to float
rookies_df.drop(['3P%'], axis=1, inplace=True)
```

```python
rookies_df.to_csv('data/cleaned_nba_hof_rookies.csv', index=False, encoding='utf-8')
```
