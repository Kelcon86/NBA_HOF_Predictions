```python
# import
import pandas as pd
```

```python
# load data
rookies_df = pd.read_csv('data/nba_hof_rookies.csv')
rookies_df.head()
```

```python
# check types
rookies_df.dtypes
```

```python
# rename some columns
rookies_df.rename(columns={'Name':'name',
                           'Hall of Fame Class':'hof_class',
                           'Year Drafted':'year_drafted',
                           '3P Made':'3PM'},
                  inplace=True)
rookies_df.head()
```

```python
rookies_df.isnull().values.any()
```

```python
rookies_df.fillna(0, inplace=True)
rookies_df.isnull().values.any()
```

```python
# Change HOF induction year with ones
rookies_df['hof_class'].values[rookies_df['hof_class'].values > 0] = 1
rookies_df.head()
```

```python
# Convert '-' strings in 3P% column to float/zeros
# source: https://stackoverflow.com/questions/33440234/pandas-how-to-replace-string-with-zero-values-in-a-dataframe-series
rookies_df['3P%'] = pd.to_numeric(rookies_df['3P%'], errors='coerce')
rookies_df['3P%'].replace('NaN',0)
```

```python
# Manual investigation of HOF players found one mislabled player.
rookies_df.loc[rookies_df['name'] == 'Bobby Jones']
```

```python
# dropped bad row
rookies_df.drop((137), inplace=True)
```

```python
# Make player names the index, then drop name feature
rookies_df.set_index(rookies_df['name'], inplace=True)
rookies_df.drop(['name'], axis=1, inplace=True)
rookies_df.head()
# remember to set names to index when loading
```
