```python
# Initial imports
import pandas as pd
import numpy as np
```

```python
# Load the nba_hof_rookies dataset.
file_path = "Resources/nba_hof_rookies.csv"
rookies_df = pd.read_csv(file_path)
rookies_df.head()
```

```python
# Check datatypes
rookies_df.dtypes
```

```python
# Clean up 3P% Column; Replace '-' with '0'
rookies_df['3P%'] = rookies_df['3P%'].replace({'-':'0'})
rookies_df.dtypes
```

```python
# Convert 3P% column to Numeric
#df["2nd"] = pd.to_numeric(df["2nd"])
rookies_df['3P%'] = pd.to_numeric(rookies_df['3P%'])
rookies_df.dtypes
```

```python
# Normalize Hall of Fame Class Column; Replace 'NaN' with 'Not Inducted'

#replace NaN values in 'points' and 'rebounds' columns with 'none'
#df[['points', 'rebounds']] = df[['points', 'rebounds']].fillna('none')

rookies_df['Hall of Fame Class'] = rookies_df['Hall of Fame Class'].fillna('Not Inducted')

rookies_df['Hall of Fame Class'] = np.where(rookies_df['Hall of Fame Class'] == "Not Inducted", "Not Inducted", "Hall of Fame Member")
rookies_df.head(50)
```

```python
# Identify incomplete rows
rookies_df.count()
```

```python
# Check datatypes
rookies_df.dtypes
```

```python
# Export to CSV
export_path = "Resources/rookies_df.csv"
rookies_df.to_csv(export_path)
```
